```
import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten
from keras.layers import Conv2D, MaxPooling2D
from keras.utils import to_categorical
from keras.preprocessing import image
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from keras.utils import to_categorical
from tqdm import tqdm
```

```
downloaded = drive.CreateFile({'id':'19XJwWgXLM2YnwNjh_zeMjPR_K1rEBtAp'}) 
downloaded.GetContentFile('train.csv')  
train = pd.read_csv('train.csv')
```

```
print(train.shape)
```

```
train['Class'][1]
```

```
train_image = []
for i in tqdm(range(train.shape[0])):
    if(train['Class'][i]=='Small'):
       img = image.load_img('Train Images/Small/'+train['Image_File'][i], target_size=(28,28,1), grayscale=False)
    else:
       img = image.load_img('Train Images/Large/'+train['Image_File'][i], target_size=(28,28,1), grayscale=False)
    img = image.img_to_array(img)
    img = img/255
    train_image.append(img)
X = np.array(train_image)
```

Replace Large with 1 and Small with 0

```
for i in tqdm(range(train.shape[0])):
    if(train['Class'][i]=='Small'):
       train['Class'][i]=0
    else:
       train['Class'][i]=1

  
```

```
y=train['Class'].to_numpy()
y=to_categorical(y)
```

```
X_train, X_val, y_train, y_val = train_test_split(X, y, random_state=42, test_size=0.2)
```

```
model = Sequential()
model.add(Conv2D(32, kernel_size=(3, 3),activation='relu',input_shape=(28,28,3)))
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(2, activation='softmax'))
```

```
model.compile(loss='categorical_crossentropy',optimizer='Adam',metrics=['accuracy'])
```

```
model.fit(X_train, y_train, epochs=10, validation_data=(X_val, y_val))
```

```
test_image = []
for i in tqdm(range(test.shape[0])):
    img = image.load_img('Test Images/'+test['Image_File'][i], target_size=(28,28,1), grayscale=False)
    img = image.img_to_array(img)
    img = img/255
    test_image.append(img)
test = np.array(test_image)
```
